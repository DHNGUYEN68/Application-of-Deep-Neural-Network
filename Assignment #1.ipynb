```python
# Programming Assignment #1, 
# Hanyu Feng ID:452106
# T81-558: Application of Deep Learning
import os
import sklearn
from sklearn.cross_validation import KFold
import pandas as pd
import numpy as np
from scipy.stats import zscore

path = "./data/"

def question1():
    print()
    print("***Question 1***")
    filename_read = os.path.join(path,"auto-mpg.csv")
    df = pd.read_csv(filename_read,na_values=['NA','?'])
    mean = df['weight'].mean()
    print("Mean weight is: {}".format(mean))
    
    
def question2():
    print()
    print("***Question 2***")
    filename_read = os.path.join(path,"auto-mpg.csv")
    filename_write = os.path.join(path,"submit-jheaton-prog1q2.csv")
    df = pd.read_csv(filename_read,na_values=['NA','?'])
    med = df['horsepower'].median()
    df['horsepower'] = df['horsepower'].fillna(med) 
    df['horsepower'] = zscore(df['horsepower'])
    df['mpg'] = zscore(df['mpg'])
    df['displacement'] = zscore(df['displacement'])
    df['weight'] = zscore(df['weight'])
    df['acceleration'] = zscore(df['acceleration'])
    mean = df['mpg'].mean()
    stdv = df['mpg'].std()
    print("MPG mean:{}".format(mean))
    print("MPG standard deviation:{}".format(stdv))
    df.to_csv(filename_write,index=False)
    

def question3():
    print()
    print("***Question 3***")
    filename_read = os.path.join(path,"auto-mpg.csv")
    df = pd.read_csv(filename_read,na_values=['NA','?'])
    filename_write = os.path.join(path,"submit-jheaton-prog1q3.csv")
    df.insert(1,'efficiency',(df['horsepower']/df['displacement']).astype(float))
    df = df.sort_values(by='efficiency',ascending=False)
    col_name = df['name']
    col_effi = df['efficiency']
    result = pd.concat([col_name,col_effi],axis=1)
    print("The most efficient is:{}".format(result['name'].iloc[0]))
    result.to_csv(filename_write,index=False)
    
def question4():
    print()
    print("***Question 4***")
    filename_read = os.path.join(path,"auto-mpg.csv")
    df = pd.read_csv(filename_read,na_values=['NA','?'])
    filename_write = os.path.join(path,"submit-jheaton-prog1q4.csv")
    mask = np.random.rand(len(df))<0.8
    trainDF = pd.DataFrame(df[mask])
    trainDF.insert(1,'set',"T")
    validateDF = pd.DataFrame(df[~mask])
    validateDF.insert(1,'set',"V")
    result = pd.concat([trainDF,validateDF],axis=0)
    print("Training DF:{}".format(len(trainDF)))
    print("Validation DF:{}".format(len(validateDF)))
    result.to_csv(filename_write,index=False)
    

def question5():
    print()
    print("***Question 5***")
    filename_read = os.path.join(path,"auto-mpg.csv")
    df = pd.read_csv(filename_read,na_values=['NA','?'])
    filename_write = os.path.join(path,"submit-jheaton-prog1q5.csv")
    df = df.reindex(np.random.permutation(df.index))
    kf = KFold(len(df), n_folds=5)
    iteration = 1
    for train_index, validate_index in kf :        
        trainDF = pd.DataFrame(df.ix[train_index,:])
        validateDF = pd.DataFrame(df.ix[validate_index])
        trainDF.insert(0,'set',"T")
        trainDF.insert(1,'iteration',iteration)
        validateDF.insert(0,'set',"V")
        validateDF.insert(1,'iteration',iteration)
        temp = pd.concat([trainDF,validateDF],axis=0)
        if iteration==1:
            result = temp
        else:
            result = pd.concat([result,temp],axis=0)
        
        print("Iteration #{}, Training Size: {}, Validation Size: {}".format(iteration,len(trainDF),len(validateDF)))
        iteration+=1
    result.to_csv(filename_write,index=False)
    
    
      

question1()
question2()
question3()
question4()
question5()
```
